```python
#Identify missing values.
pd.DataFrame(original_data.isnull().sum(), columns=['Missing Values'])
```

##2.3 Descriptive Stats

```python
#View metadata
original_data.info()
```

- The data consisits of 20 columns that are a combination of numeric and categorical types
- There are ~1m rows in the dataset

```python
#Convert "numeric object" columns to numeric data types in order to perform descriptive stats

#Copy original data
data = original_data.copy(deep=True)

#Regex to remove unwanted characters
currency_chars_to_remove = r'[$,()£€AZaz\s]+'

#Columns to convert
columns_to_convert = ['Total Revenue', 'Unit Cost', 'Profit']

#Loop through columns to perform conversion
for col in columns_to_convert:

    #Convert to string, clean unwanted characters
    org_string = data[col].astype(str)
    clean_string = org_string.str.replace(currency_chars_to_remove, '', regex=True)

    #Convert to numeric
    numeric_values = pd.to_numeric(clean_string, errors='coerce')

    #Identify negatives by original formatting
    negative_mask = org_string.str.contains(r'\(', regex=True) & numeric_values.notna()

    #Apply negation where appropriate
    numeric_values[negative_mask] *= -1

    #Update DataFrame
    data[col] = numeric_values

#Summary statistics on numeric columns
data[['Quantity', 'Total Revenue', 'Unit Cost', 'Profit', 'Days to Delivery']].describe()
```

- Identified columns Quantity, Total Revenue, Unit Cost, Profit & Days to delivery successfully converted to numeric datatypes
- Descriptive stats on these columns show large ranges across each feature and not necessarily symetrically distributed

##2.6 Create new features & aggregate the data into one customer per row

```python
#Import libraries
from datetime import date, datetime

#Calc/Create Frequency feature
data['Frequency'] = data.groupby('Customer ID')['Order ID'].transform('nunique')

#Get Today's date for recency calc
today = pd.Timestamp('now')

#Calc/Create Recency feature
data['Recency'] = (pd.to_datetime(today) - pd.to_datetime(data['Delivery_Date'])).dt.days

#Convert the DOB to datetime
data['DOB'] = pd.to_datetime(data['Customer_BirthDate'], format= '%d%b%Y')
#Calc/Create Age feature
data['Age'] = (today - data['DOB']).dt.days // 365

#Rename the Total Revenue and Unit Costs columns
data.rename(columns={'Total Revenue' : 'CLV', 'Unit Cost' : 'Average_Unit_Cost'}, inplace=True)

#Select aggregations per feature
agg_operations = {'Frequency': 'max',
                  'Recency': 'max',
                  'Average_Unit_Cost': 'mean',
                  'CLV': 'sum',
                  'Age': 'first'}

#Perform aggregation and create new dataframe
data_aggr = data.groupby('Customer ID').agg(agg_operations).reset_index()

#Check new aggregated dataframe
data_aggr.head()
```

- The aggregation results in 68,300 records of numeric data in a new dataframe containing the 5 new features per individual customer

##5. k-Means clustering

###5.1 Elbow Plots

```python
# Import k-Means function.
from sklearn.cluster import KMeans

#Create list to store each of the WCSS values per cluster.
wcss=[]

#Loop though 2 to 10 clusters and applying k-means
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0, n_init = 10)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)

#Plot scartter plot of clusters vs their WCSS
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

- The Elbow plot is not conclusive and it would see the optimal cluster number is between 3 and 5

###5.2 Silhouette Plots

####5.2.1 Calc and compare avg silhouette scores for different clusters

```python
#import required libraries
from sklearn.metrics import silhouette_score

#Initialize lists to store results
n_clusters_list = []
silhouette_avg_list = []

#Loop though 2 to 10 clusters and applying k-means
for n_clusters in range(2, 11):

    #Initialize the clusterer with n_clusters value and a random generator
    #seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, init='k-means++', random_state=10, n_init = 10)
    cluster_labels = clusterer.fit_predict(scaled_df)

    #Calculate average silhouette score per cluster
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    #Save cluster and associated silhouette scores for plotting
    n_clusters_list.append(n_clusters)
    silhouette_avg_list.append(silhouette_avg)

#Create the scatter plot of cluster numbers vs their silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(n_clusters_list, silhouette_avg_list, marker='o', linestyle='-')

plt.title('Average Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters (n_clusters)')
plt.ylabel('Average Silhouette Score')
plt.xticks(n_clusters_list)

plt.grid(True)

plt.show()
```

- The graph shows 5 clusters to have the highest silhouette score

####5.2.2 Plot the silhouette graphs

```python
#Import required libraries
import matplotlib.cm as cm
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Create a 3x2 grid of subplots (3 rows, 2 columns)
fig, axes = plt.subplots(3, 2, figsize=(14, 18))
axes = axes.flatten()

#Loop through 2 to 6 clusters to plot silhouette graphs
for i, n_clusters in enumerate(range(2, 7)):

    #Fit k-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = kmeans.fit_predict(scaled_df)

    #Calc Silhouette scores
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)
    sample_silhouette_values = silhouette_samples(scaled_df, cluster_labels)

    #Set plot axes
    ax = axes[i]
    ax.set_xlim([-0.2, 1])
    ax.set_ylim([0, len(scaled_df) + (n_clusters + 1) * 10])

    y_lower = 10
    for j in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == j]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(j) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)

        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(j))
        y_lower = y_upper + 10

    #Add vertical average line
    line = ax.axvline(x=silhouette_avg, color="red", linestyle="--", label=f"Mean: {silhouette_avg:.4f}")

    #Add legend with mean value
    ax.legend(loc="upper right", fontsize=10)

    ax.set_title(f"Silhouette Plot for {n_clusters} Clusters")
    ax.set_xlabel("The silhouette coefficient values")
    ax.set_ylabel("Cluster label")
    ax.set_yticks([])
    ax.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

# Remove unused subplots if fewer than 6
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

- No one plot/cluster seems to be conclusive "clean" with cross-over/overlap of data points between most clusters across each plot
- However, clusters 3 and 4 seem to present less of a overlap with 3 less than 4

```python
def plot_scatter_plot (df, redu_method,cl):

  #Create the scatter plot
  plt.figure(figsize=(10, 8))
  sns.scatterplot(data=df, x='PC1', y='PC2', hue='Cluster', palette='viridis', legend='full')

  plt.title(f"{redu_method} Plot with Clusters = {cl}", fontsize=16)
  plt.xlabel('Principal Component 1')
  plt.ylabel('Principal Component 2')

  plt.grid(True)
  plt.show()
  return
```

###9.1 PCA Reduction
